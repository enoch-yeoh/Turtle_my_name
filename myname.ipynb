```python
import turtle
#trying to write my name on Turtle"
# Screen setup
screen = turtle.Screen()
screen.bgcolor("black")

# Turtle setup
pen = turtle.Turtle()
pen.speed(0)
pen.color("white")
pen.penup()
pen.hideturtle()  # Hide the turtle cursor for cleaner drawing

def draw_E(x, y, size):
    # Vertical line
    pen.penup()
    pen.goto(x, y)
    pen.pendown()
    pen.goto(x, y + size)
    
    # Three horizontal lines
    pen.goto(x + size/2, y + size)  # Top
    
    pen.penup()
    pen.goto(x, y + size/2)
    pen.pendown()
    pen.goto(x + size/2, y + size/2)  # Middle
    
    pen.penup()
    pen.goto(x, y)
    pen.pendown()
    pen.goto(x + size/2, y)  # Bottom

def draw_N(x, y, size):
    # Left vertical line
    pen.penup()
    pen.goto(x, y)
    pen.pendown()
    pen.goto(x, y + size)
    
    # Diagonal line
    pen.goto(x + size/2, y)
    
    # Right vertical line
    pen.goto(x + size/2, y + size)

def draw_O(x, y, size):
    pen.penup()
    pen.goto(x + size/4, y)  # Bottom center of O
    pen.pendown()
    pen.setheading(90)  # Point up
    
    # Draw left half
    pen.circle(size/4, 180)
    
    # Continue to draw right half
    pen.circle(size/4, 180)

def draw_C(x, y, size):
    pen.penup()
    pen.goto(x + size/2, y + size/4)  # Right middle
    pen.setheading(90)  # Point up
    pen.pendown()
    
    # Draw 3/4 of a circle
    pen.circle(size/4, 180)
    pen.circle(size/4, 90)

def draw_H(x, y, size):
    # Left vertical line
    pen.penup()
    pen.goto(x, y)
    pen.pendown()
    pen.goto(x, y + size)
    
    # Horizontal line
    pen.penup()
    pen.goto(x, y + size/2)
    pen.pendown()
    pen.goto(x + size/2, y + size/2)
    
    # Right vertical line
    pen.penup()
    pen.goto(x + size/2, y)
    pen.pendown()
    pen.goto(x + size/2, y + size)

# Starting position
x = -200
y = -25
size = 50
spacing = 60

# Draw each letter
draw_E(x, y, size)
x += spacing
draw_N(x, y, size)
x += spacing
draw_O(x, y, size)
x += spacing
draw_C(x, y, size)
x += spacing
draw_H(x, y, size)

# Keep the window open
screen.mainloop()
```
